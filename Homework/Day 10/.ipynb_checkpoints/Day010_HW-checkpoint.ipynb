## requests + BeautifulSoup


```python
import requests

r = requests.get('https://medium.com/')
resp = r.text
resp
```

```python
import json
json.loads(resp)
```

```python
from bs4 import BeautifulSoup
soup = BeautifulSoup(r.text, "html5lib")
print(soup)
```

## Grab + PyQuery

```python
pip install grab
```

```python
from grab import Grab
g = Grab()
resp = g.go('https://medium.com/')
resp.body
```

```python
pip install pyquery
```

```python
from pyquery import PyQuery as pq
doc = pq(resp.body)
title = doc('title')
print(type(title), title.text())
```

```python
div = doc('div')
print(div.text())
```

```python
h1 = doc('h1')
print(h1.text())
```

```python
h2 = doc('h2')
print(h2.text())
```
