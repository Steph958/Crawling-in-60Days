### ① 取出今天所有的發文

```python
import requests
from bs4 import BeautifulSoup

url = 'https://www.ettoday.net/news/news-list.htm'
r = requests.get(url)
resp = r.text

print(resp)
```

```python
soup = BeautifulSoup(resp, "html5lib")

for i in soup.find(class_ = "part_list_2").find_all('h3'):
    print(i.find(class_="date").text, i.find_all('a')[-1].text)
```

### ② 如果想要依照類別分類，怎麼儲存會比較好？

```python
url = 'https://www.ettoday.net/news/news-list.htm'

r = requests.get(url)
soup = BeautifulSoup(r.text, "html5lib")

categories = {}

for d in soup.find(class_="part_list_2").find_all('h3'):
    date = d.find(class_="date").text
    title = d.find_all('a')[-1].text
    tag = d.find(class_="tag").text
    #tag = d.select('.tag')[0].text
    
    categories.setdefault(tag, [])
    categories[tag].append({'title':title, 'date':date})
    
    
categories
```

### ③ 哪一個類別的文章最多

```python
for i in categories:
    print(i, len(categories[i]))
```

# 用字典儲存:

```python
data = []

for i in categories:
    d = {}
    d['tag'] = i
    d['count'] = len(categories[i])
    data.append(d)
    
data
```

```python
for d in sorted(data, key=lambda d: d['count'], reverse = True):
    print(d)
```

# 用資料框儲存:

```python
import pandas as pd

data_2 = pd.DataFrame(data)
data_2
```

```python
data_2.sort_values(by='count', ascending = False)
```
