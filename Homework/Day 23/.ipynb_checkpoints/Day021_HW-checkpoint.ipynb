## 1. 根據範例，取出今天所有的新聞

```python
from selenium import webdriver
from bs4 import BeautifulSoup

browser = webdriver.Chrome(executable_path='./chromedriver.exe')

browser.get("https://www.ettoday.net/news/news-list.htm")
```

```python
#得到資料
html_source = browser.page_source
soup = BeautifulSoup(html_source, "html5lib")
#print(soup)
#print(soup.find_all(class_="part_list_2")[-1])
#print(soup.find_all(class_="part_list_2")[-1].find_all('h3')[-1])
#print(soup.find_all(class_="part_list_2")[-1].find_all('h3')[-1].find(class_="date"))
print(soup.find_all(class_="part_list_2")[-1].find_all('h3')[-1].find(class_="date").text)
```

```python
# 取得昨天時間
from datetime import datetime, timedelta
date_time = datetime.now() - timedelta(1)
date_time = date_time.strftime("%Y/%m/%d")
print("取得昨天時間：", date_time)
```

```python
# 利用 selenium 持續執行下滑的動作，直到取得「昨天時間」的新聞，表示今天的已經取得完畢
import time

while True:
    browser.execute_script("window.scrollTo(0, 100000);")
    time.sleep(2)
    
    html_source = browser.page_source
    soup = BeautifulSoup(html_source, "html5lib")
    
    print(soup.find_all(class_="part_list_2")[-1].find_all('h3')[-1].find(class_="date").text) #最後(最早)的一筆資料的時間
    
    if date_time in soup.find_all(class_="part_list_2")[-1].find_all('h3')[-1].find(class_="date").text:
        print('==== STOP ====')
        break
```

```python
#print(soup.find(class_="part_list_2").find_all('h3'))
#print(soup.find(class_="part_list_2").find('h3').find_all(class_="date"))
print(soup.find(class_="part_list_2").find('h3').find(class_="date").text)

#註:在搜尋鏈中，最後才用find_all()，前面都用find()
```

```python
# 取得資料
html_source = browser.page_source
soup = BeautifulSoup(html_source, "html5lib")

for d in soup.find(class_="part_list_2").find_all('h3'):
    if date_time in d.find(class_="date").text :   #拿昨天的時間，比對今天最早一筆資料的時間
        break
    print(d.find(class_="date").text, d.find_all('a')[-1].text)
```

```python
# 取得資料（整理成好的格式）
news = []

for d in soup.find(class_="part_list_2").find_all('h3'):
    if date_time in d.find(class_="date").text :
        break
    news.append({
        '時間': d.find(class_="date").text,
        '標題': d.find_all('a')[-1].text
    })
    
print(news)

import pandas as pd
pd.DataFrame(news)
```

```python
# 關閉瀏覽器

browser.close()
```

## 2. 取出現在時間兩小時內的新聞

```python
# 取得昨天時間
from datetime import datetime, timedelta
date_time = datetime.now() - timedelta(1)
print(date_time)

date_time = date_time.strftime("%Y/%m/%d") #時間格式轉換
print("取得昨天時間：", date_time)
```

```python
# 取得兩小時內的時間
from datetime import datetime, timedelta

two_hours_ago_time = datetime.now() - timedelta(hours=2)
print(two_hours_ago_time)

# 取得資料（整理成好的格式）
news = []

for d in soup.find(class_="part_list_2").find_all('h3'):
    # 比兩小時前早的話就結束
    if datetime.strptime(d.find(class_="date").text, '%Y/%m/%d %H:%M') < two_hours_ago_time: #爬蟲取出的時間是文字格式，還需轉換才能比對
        break
    
    news.append({
        '時間': d.find(class_="date").text,
        '標題': d.find_all('a')[-1].text
    })

import pandas as pd
pd.DataFrame(news)
```

## 3. 根據範例，取出三天前下午三點到五點的新聞

```python
from selenium import webdriver
from bs4 import BeautifulSoup

browser = webdriver.Chrome(executable_path='./chromedriver.exe')

browser.get("https://www.ettoday.net/news/news-list.htm")
```

```python
# 取得三天前下午三點到五點的時間
from datetime import datetime, timedelta

date_time = datetime.now() - timedelta(3)
print("取得三天前日期時間：", date_time)

date_time_month = date_time.strftime("%m")
date_time_day = date_time.strftime("%d")
print("取得三天前時間（月）：", date_time_month)
print("取得三天前時間（日）：", date_time_day)


three_days_15_clock = datetime.strptime(date_time.strftime("%Y/%m/%d") + ' 15', '%Y/%m/%d %H')
three_days_17_clock = datetime.strptime(date_time.strftime("%Y/%m/%d") + ' 17', '%Y/%m/%d %H')
print("取得三天前下午三點：", three_days_15_clock)
print("取得三天前下午五點：", three_days_17_clock)
```

```python
# 選擇且點選日期
from selenium.webdriver.support.ui import Select

selM = Select(browser.find_element_by_id("selM"))
selM.select_by_value(date_time_month[1])   #月份下拉選單中選6月

selD = Select(browser.find_element_by_id("selD"))
selD.select_by_value(date_time_day[1])    #日期下拉選單中選6號

browser.find_element_by_id('button').click()   #按下查詢按鈕
```

```python
# 利用 selenium 持續執行下滑的動作，直到取得要求時間內的新聞，表示取得完畢
import time

while True:
    browser.execute_script("window.scrollTo(0, 100000);")
    time.sleep(1)
    
    html_source = browser.page_source
    soup = BeautifulSoup(html_source, "html5lib")
    
    new_date = soup.find_all(class_="part_list_2")[-1].find_all('h3')[-1].find(class_="date").text
    print(new_date)
    
    if datetime.strptime(new_date, '%Y/%m/%d %H:%M') < three_days_15_clock:
        print('==== STOP ====')
        break
```

```python
# 取得資料（整理成好的格式）
news = []

#要抓下午3點到5點的資料
#資料是由後往前的跑迴圈

for d in soup.find(class_="part_list_2").find_all('h3'):
    # 比限定的下午3點還早的資料 >> 就結束
    if datetime.strptime(d.find(class_="date").text, '%Y/%m/%d %H:%M') < three_days_15_clock:
        print('==== STOP ====')
        break
    #在限定的下午5點之前的資料 >> 印出來
    if datetime.strptime(d.find(class_="date").text, '%Y/%m/%d %H:%M') < three_days_17_clock:
        news.append({
            '時間': d.find(class_="date").text,
            '標題': d.find_all('a')[-1].text
        })

import pandas as pd
pd.DataFrame(news)
```

```python
# 關閉瀏覽器

browser.close()
```
