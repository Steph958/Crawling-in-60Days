```python
import requests

r = requests.get(' https://www.dcard.tw/_api/forums/pet/posts?popular=true')
resp = r.text

resp
```

```python
import json
data = json.loads(resp)

data
```

```python
# 1. 這個 API 一次會回傳幾筆資料？每一筆資料包含哪些欄位？

ids = []

for list_item in data:
    dic = list_item
    print(dic['id'])
    ids.append(dic['id'])

print('================================')
print('共有' + str(len(ids)) +  "筆資料")
```

```python
# 2. 取出每一筆資料的「標題」、「貼文時間」、「留言人數」、「按讚人數」

id = 0

for i in data:
    id+=1
    print('筆數:' + str(id))
    print('標題:' + i['title'])
    print('貼文時間:' + i['createdAt'])
    print('留言數:' + str(i['commentCount']))
    print('按讚數:' + str(i['likeCount']))
    print('================================================')
    
```
