```python
import requests
import re
from bs4 import BeautifulSoup
import time
```

先試看看用單線程爬yahoo電影清單

```python
start=time.time()

page = 0

for i in range(1,5):
    url = 'https://movies.yahoo.com.tw/movie_intheaters.html'+'?page='+str(i)
    response = requests.get(url)
    soup = BeautifulSoup(response.text, 'xml')
    movie_list = soup.find('div', class_="top_change").find_all('option')
    
    print('=====================================================================================================')
    print('Following is page'+str(i))
    print('=====================================================================================================')
        
    for j in movie_list:
        print(j)
        
        
done=time.time()
print('=====================================================================================================')
print('Done!')
print('total_time:',done-start)
    
```

接著試看看多線程:

```python
import _thread
import time
import requests
from bs4 import BeautifulSoup
```

```python
outcome=[]
```

```python
def job(url):
    response = requests.get(url)
    soup = BeautifulSoup(response.text, 'xml')
    movie_list = soup.find('div', class_="top_change").find_all('option')
    outcome.append(movie_list)
```

```python
start=time.time()

page = 0

for i in range(1,5):
    url = 'https://movies.yahoo.com.tw/movie_intheaters.html'+'?page='+str(i)

    _thread.start_new_thread(job, (url, ) )

done=time.time()
print('=====================================================================================================')
print('Done!')
print('total_time:',done-start)
```
