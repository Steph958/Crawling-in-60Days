```python
from selenium import webdriver
from bs4 import BeautifulSoup
import time
import pandas as pd

URL='https://www.cupoy.com/newsfeed/topstory'
```

```python
# 打開瀏覽器, 進入欲爬取網頁
browser = webdriver.Chrome(executable_path='../chromedriver')
browser.get("URL")
```

```python
start=time.time()
  
last_height = browser.execute_script("window.scrollTo(0, document.body.scrollHeight);")
title_num=0
data_info=[]

while True:
    
    browser.execute_script("window.scrollTo(0, document.body.scrollHeight);")
    time.sleep(3)
    new_height = browser.execute_script("return document.body.scrollHeight")


    html_source = browser.page_source
    soup = BeautifulSoup(html_source, "lxml")
    T = soup.find_all('a',class_="sc-jxGEyO")

    for i in T:
        
        data={}
        #標題
        data['Title']=i['title']
        #文章連結網頁
        data['URL']=i['href']
        #原始網頁
        data['Web']=i['href'].split('/')[2]
        
        if data not in data_info:
            data_info.append(data)
            title_num+=1
            
    print('now_num:',title_num)
            
    if title_num >= 500:
        print('==== STOP ====')
        print('final_num:',title_num)
        break
    
    last_height = new_height
    
browser.close()
end=time.time()
print('time:',end-start)
```

```python
import requests
from bs4 import BeautifulSoup
import re

headers={'user-agent':'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/94.0.4606.61 Safari/537.36 Edg/94.0.992.37'}

multi_content=[]

for i in news_url:
    
    single_content='' #存每個url讀取到的最終結果
    
    response = requests.get(i, headers=headers)  
    response.encoding='utf-8'
    soup = BeautifulSoup(response.text, 'html.parser')
        
    paragraphs = []     #第一手爬到的資料
            
    if soup.find('p'):                                         # 新聞內文有些在 section 標籤，有些在 p 標籤
        paragraphs.append(soup.find_all('p'))                 
                                                                             
    if soup.find('section'):
        paragraphs.append(soup.find_all('section'))

    for pars in paragraphs:                                    
        for par in pars:
            text = re.sub('[\W]+', ' ', par.text)                   # 找出特殊字元 (\n,\t,...等) 替換成空白      
            single_content += text
                
        if single_content != '' and type(single_content) == str:        # 確認抓取到的內容是否為空，否則再重新抓一次
            print('Successfully get information.')
            
        else:
            print('Something Goes Wrong.')
            
    multi_content.append(single_content)
    
```
