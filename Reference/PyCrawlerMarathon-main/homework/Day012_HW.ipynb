### ① 取出今天所有的發文

```python
'''
Your Code
'''
import requests
from bs4 import BeautifulSoup

url = 'https://www.ettoday.net/news/news-list.htm'
r = requests.get(url)

soup = BeautifulSoup(r.text)

#print(soup.prettify())
```

```python
for d in soup.find(class_="part_list_2").find_all('h3'):
    print(d.find(class_="date").text,d.find_all('em')[-1].text,d.find_all('a')[-1].text)
```

### ② 如果想要依照類別分類，怎麼儲存會比較好？

```python
'''
Your Code
'''
import pandas as pd


label=[]
date=[]
title=[]

for d in soup.find(class_="part_list_2").find_all('h3'):
    label.append(d.find_all('em')[-1].text)
    date.append(d.find(class_="date").text)
    title.append(d.find_all('a')[-1].text)

dict1 = {'class':label,
         'date':date,
         'title':title
        }
   
select_df = pd.DataFrame(dict1)

#依據class(分類)和date進行排序
D = select_df.sort_values(by=['class','date'],ascending = False)
print(D)
```

### ③ 哪一個類別的文章最多

```python
'''
Your Code
'''
#建立類別字典
class_dict={}

for d in soup.find(class_="part_list_2").find_all('h3'):
    
    if (d.find_all('em')[-1].text) not in class_dict:
        class_dict[(d.find_all('em')[-1].text)]=1
    else:
        class_dict[(d.find_all('em')[-1].text)]+=1

print("此網頁有{}則新聞".format(sum(list(class_dict.values()))))
print('\n')
print(class_dict)
```

```python
print("文章最多的類別：",max(class_dict,key=class_dict.get))
```
