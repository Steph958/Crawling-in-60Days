## 1. 取出今天所有的新聞

```python
from selenium import webdriver
from bs4 import BeautifulSoup
import time
import datetime as dt
```

```python
date_time = dt.datetime.now() - dt.timedelta(1)
date_time = date_time.strftime("%Y/%m/%d")
print("取得昨天時間：", date_time)
```

```python
#此滾輪方式適合用一整天的
#每次滾至最底部，再將現在位置設為下次滾動的初始位置
browser = webdriver.Chrome('/Users/andy/Desktop/chromedriver')

browser.get("https://www.ettoday.net/news/news-list.htm")
last_height = browser.execute_script("window.scrollTo(0, document.body.scrollHeight);")


while True:
    
    browser.execute_script("window.scrollTo(0, document.body.scrollHeight);")
    time.sleep(2)
    new_height = browser.execute_script("return document.body.scrollHeight")


    html_source = browser.page_source
    soup = BeautifulSoup(html_source, "lxml")

    print(soup.find_all(class_="part_list_2")[-1].find_all('h3')[-1].find(class_='date').text)

    if date_time in soup.find_all(class_="part_list_2")[-1].find_all('h3')[-1].find(class_="date").text:
        print('==== STOP ====')
        break
    last_height = new_height
```

```python
#滾完再取資料
html_source = browser.page_source
soup = BeautifulSoup(html_source, "lxml")

for d in soup.find(class_="part_list_2").find_all('h3'):
    if date_time in d.find(class_="date").text :
        break
    print(d.find(class_="date").text, d.find_all('a')[-1].text)
```

```python
browser.quit()
```

## 2. 取出現在時間兩小時內的新聞

```python
start_time = dt.datetime.now()
print(start_time)
delta = dt.timedelta(hours=2)


browser = webdriver.Chrome('/Users/andy/Desktop/chromedriver')

browser.get("https://www.ettoday.net/news/news-list.htm")

end_time = start_time - delta

#只取短短兩小時且離現在時間點很近，故用此方法也行
for i in range(10):
    time.sleep(2)
    browser.execute_script("window.scrollBy(0, 10000);")

html_source = browser.page_source
soup = BeautifulSoup(html_source, "lxml")

for d in soup.find(class_="part_list_2").find_all('h3'):
    
    T = dt.datetime.strptime(d.find(class_='date').text, '%Y/%m/%d %H:%M')
    
    if T < end_time:
        break
    
    print(d.find(class_="date").text, d.find_all('a')[-1].text)

browser.quit()
```

## 3. 根據範例，取出三天前下午三點到五點的新聞

```python
today = dt.datetime.now()
the_day = today-dt.timedelta(days=3)
the_day_time = dt.datetime(the_day.year, the_day.month, the_day.day, 15, 0, 0)
delta = dt.timedelta(hours=2)

day_time = the_day_time.strftime("%Y/%m/%d %H:%S")

browser = webdriver.Chrome('/Users/andy/Desktop/chromedriver')
url = "https://www.ettoday.net/news/news-list.htm"
new_url = url.replace('news-list', 'news-list-'+str(the_day.year)+'-'+str(the_day.month)+'-'+str(the_day.day)+'-0')

browser.get(new_url)
last_height = browser.execute_script("window.scrollTo(0, document.body.scrollHeight);")

#先用滾輪到下午3.
while True:
    
    browser.execute_script("window.scrollTo(0, document.body.scrollHeight);")
    time.sleep(2)
    new_height = browser.execute_script("return document.body.scrollHeight")


    html_source = browser.page_source
    soup = BeautifulSoup(html_source, "lxml")

    print(soup.find_all(class_="part_list_2")[-1].find_all('h3')[-1].find(class_='date').text)

    if day_time in soup.find_all(class_="part_list_2")[-1].find_all('h3')[-1].find(class_="date").text:
        print('==== STOP ====')
        break
    last_height = new_height
```

```python
#取下午3.~5.資料    
html_source = browser.page_source
soup = BeautifulSoup(html_source, "lxml")
                     
for d in soup.find(class_="part_list_2").find_all('h3'):
    
    T = dt.datetime.strptime(d.find(class_='date').text, '%Y/%m/%d %H:%M')
    
    
    if T < the_day_time:
        break
    elif T < (the_day_time+delta):
        print(d.find(class_="date").text, d.find_all('a')[-1].text)
    else:
        pass
    
browser.quit()
```
