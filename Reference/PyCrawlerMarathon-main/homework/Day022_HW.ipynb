### 2. 剛剛的做法只能取出第一頁，試著取出所有頁面的資料。

```python
from selenium import webdriver
from selenium.webdriver.support.ui import Select
from bs4 import BeautifulSoup

browser = webdriver.Chrome('/Users/andy/Desktop/chromedriver')
browser.get("https://hahow.in/courses")
```

```python
# 取得資料（等瀏覽器資料出現後才執行）
html_source = browser.page_source
soup = BeautifulSoup(html_source, 'lxml')
```

```python
# 1. 先取出全部有幾頁
soup.find('div',class_='pagination-container').find_all('a')
browser.close()
#25
```

```python
# 2. 每頁都取出資料放入 courses
import time
browser = webdriver.Chrome('/Users/andy/Desktop/chromedriver')
url = "https://hahow.in/courses"
for p in range(25):
    courses = []
    if p==0:
        browser.get(url)
    
    else:
        url_page=url.replace('courses','courses'+'?page='+str(p+1))
        #print(url_page)
        browser.get(url_page)
        time.sleep(3)
    
    html_source = browser.page_source
    soup = BeautifulSoup(html_source, 'lxml')
    
    for i in soup.find_all('h4',class_='title'):
        courses.append(i.text)
    
    print('Page{}:'.format(p+1))
    print(courses)

browser.close()
```
