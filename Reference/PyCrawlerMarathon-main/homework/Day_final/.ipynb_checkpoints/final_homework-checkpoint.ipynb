# 動態網頁爬500篇文章

```python
URL=' https://www.cupoy.com/newsfeed/topstory'
```

```python
start=time.time()
browser = webdriver.Chrome('/Users/andy/Desktop/chromedriver')
browser.get(URL)  
time.sleep(5)  

last_height = browser.execute_script("window.scrollTo(0, document.body.scrollHeight);")
title_num=0
data_info=[]

while True:
    
    browser.execute_script("window.scrollTo(0, document.body.scrollHeight);")
    time.sleep(3)
    new_height = browser.execute_script("return document.body.scrollHeight")


    html_source = browser.page_source
    soup = BeautifulSoup(html_source, "lxml")
    T = soup.find_all('a',class_="sc-jxGEyO")

    for i in T:
        
        data={}
        #標題
        data['Title']=i['title']
        #文章連結網頁
        data['URL']=i['href']
        #原始網頁
        data['Web']=i['href'].split('/')[2]
        
        if data not in data_info:
            data_info.append(data)
            title_num+=1
            
    print('now_num:',title_num)
            
    if title_num >= 500:
        print('==== STOP ====')
        print('final_num:',title_num)
        break
    
    last_height = new_height
    
browser.close()
end=time.time()
print('time:',end-start)
```

```python
DATA = pd.DataFrame(data_info)
DATA.drop(DATA.index[500:],inplace=True)
DATA
```

# 分群：依出現次數分 (1) >20 (2)10 ~20 (3)5 ~10 (4) <5

```python
def trans(web,num):
    if num<=5:
        return 'lower_than_5'
    elif num>5 and num<=10:
        return '5-10'
    elif num>10 and num<=20:
        return '10-20'
    else:
        return 'outstrip_20'

appear_time=[]
for i in range(len(web)):
    appear_time.append(trans(web['web'][i],web['num'][i]))
web['Appear_Time']=appear_time
web_copy1=web.copy()
web_copy2=web.copy()
web_copy3=web.copy()
web_copy4=web.copy()
web_copy1
```

```python
count1=count2=count3=count4=0
n=[]
for i in range(len(web)):
    
    #紀錄'Appear_Time'變化成下一個時的第一時間index位置
    if web.iloc[i-1]['Appear_Time']!=web.iloc[i]['Appear_Time'] and web.iloc[i]['Appear_Time']==web.iloc[i+1]['Appear_Time']:
        n.append(i)
    
    
    if web.iloc[i]['Appear_Time']=='outstrip_20': #計算'outstrip_20'數量
        count1+=web.iloc[i]['num']
    
    elif web.iloc[i]['Appear_Time']=='10-20':     #計算'10-20'數量
        count2+=web.iloc[i]['num']
    
    elif web.iloc[i]['Appear_Time']=='5-10':      #計算'5-10'數量
        count3+=web.iloc[i]['num']
    
    else: 
        count4+=web.iloc[i]['num']                #計算'lower_than_5'數量
print(n)        

d={'outstrip_20':count1,
   '10-20':count2,
   '5-10':count3,
   'lower_than_5':count4
  }

web_copy= pd.DataFrame(list(d.items()),columns=['Appear_Time','count'])
web_copy
```

```python

def make_autopct(values):
    def my_autopct(pct):
        total = sum(values)
        val = int(round(pct*total/100.0))
        # 同时显示数值和占比的饼图
        return '{p:.2f}%  ({v:d})'.format(p=pct,v=val)
    return my_autopct


color = sns.color_palette("Blues_r")
plt.figure(figsize=(20, 10))
plt.pie(web_copy['count'], labels=web_copy['Appear_Time'], autopct=make_autopct(web_copy['count']),colors=color,textprops ={'fontsize':18})
plt.legend(fontsize=12,loc=u'upper left')
plt.title(u'News source website distribution',fontsize=20)
plt.show()
```

# 爬取500篇文章並存為txt

```python
header={'user-agent':'Mozilla/5.0 (Macintosh; Intel Mac OS X 10_12_6) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/89.0.4389.90 Safari/537.36'}

start=time.time()
for i in range(len(url_list)):

    single_content=''

    r=requests.get(url_list[i],headers=header)
    soup=BeautifulSoup(r.text,'lxml')

    if soup.find('section'):
        for t in soup.find_all('section'):
            
            text = re.sub('[\W]+', ' ', t.text)                       
            single_content += text
                

    if soup.find('p'):
        for t in soup.find_all('p'):
            
                text = re.sub('[\W]+', ' ', t.text)                 
                single_content += text
    
    f = open("content.txt", "a+", encoding='utf-8')
    f.write( single_content + "\n" )  
    f.write( "-"*90 + "\n" )
    f.close()
    
    time.sleep(2)
    if i%10==0:
        print("Now is at:",i)

end=time.time()
print('-'*15)
print("Done:",end-start)
```

# 開啟txt檔

```python
result=[]
f=open("content.txt",'r')        

for line in f:
    if line !='-'*90+'\n':
        result.append(line.strip('\n').split(' '))
```

# 去除停用詞

```python
import jieba
import jieba.analyse
import time

# 移除停用詞
stopwords=[]

#使用從https://github.com/fxsjy/jieba抓來並修改的停用詞表
with open('stop_words.txt', 'r', encoding='utf-8') as f: 
    for data in f.readlines():
        data = data.strip()
        stopwords.append(data)

# jieba 分詞
start=time.time()

C=[]

#分詞後去除停用詞, 然後再重組回文章
for n in range(len(result)):
    remained_news = []
    for i in result[n]: 
        seg = jieba.cut(i, cut_all=False)               
        for e in seg:
            if e not in stopwords and e!=' ':
                e = filter(lambda ch: ch not in '0123456789', e)
                remained_news.append(''.join(list(e)))
    C.append(''.join(remained_news))

end=time.time()
print('Done:',end-start)
```

# 尋找每篇文章內的前20關鍵字

```python
# 關鍵字 tfidf
start=time.time()
keywords = []
for n in C:
    keywords.append(jieba.analyse.extract_tags(n, topK=20, withWeight=False))

end=time.time()
print('Done:',end-start)

for j in keywords:
    print(j)
```

```python
allkeys=''
for n in keywords:                                         
    for i in n:
        allkeys+=i

#print(allkeys)
print(jieba.analyse.extract_tags(allkeys, topK=20, withWeight=False))
```

```python
def wordcloud(content):                               # 把關鍵字轉文字雲
    from wordcloud import WordCloud
    import jieba
    import matplotlib.pyplot as plt
    from PIL import Image 
    import numpy as np
    
    
    
    words = jieba.cut(content, cut_all=False)
    
    all_words = ''
    for word in words:
        all_words += ' '+word
    
    

    wcloud = WordCloud(width=500, height=500, background_color='white', 
                       min_font_size=8,font_path='/Library/Fonts/PingFang.ttc').generate(all_words)
    plt.figure(figsize=(20, 15), facecolor=None)
    plt.imshow(wcloud)
    plt.axis('off')
    plt.tight_layout(pad=0)
    
    plt.show()



wordcloud(allkeys)
```
