### 3. 觀察一下知乎回來的資料好像有點怪怪的，該怎麼解決？

status_code 是 400，被拒絕存取，加上 headers 包進 requests.get 裡

```python
import requests
url = 'https://www.zhihu.com/explore'

'''
Your Code Here
'''

headers = {'User-Agent':'Mozilla/5.0'}
r = requests.get(url, headers=headers)

r.encoding = 'utf-8'
print(r.text[0:1000])
```
