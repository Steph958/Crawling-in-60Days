### ① 取出今天所有的發文

```python
'''
Your Code
'''
import requests
from bs4 import BeautifulSoup

url = 'https://www.ettoday.net/news/news-list.htm'
res = requests.get(url)

soup = BeautifulSoup(res.text, 'html5lib')

news_list = soup.find('div', {'class':'part_list_2'}).find_all('h3')

for new in news_list:
    print(new.span.text, new.a.text)
```

### ② 如果想要依照類別分類，怎麼儲存會比較好？

```python
'''
Your Code
'''

categories = []
news = []


for new in news_list:
    category = new.em.text
    if category not in categories:
        categories.append(category)
        news.append([[new.span.text, new.a.text]])
    else:
        index = categories.index(category)
        news[index].append([new.span.text, new.a.text])

print(categories)

for i, new in enumerate(news):
    print('類別: ', categories[i])
    for j, n in enumerate(new, start=1):
        print('  {}.'.format(j), n[0], n[-1])
```

### ③ 哪一個類別的文章最多

```python
'''
Your Code
'''

maxi = 0
for i, new in enumerate(news):
    if maxi < len(new):
        maxi = len(new)
        max_index = i

print('文章最多類別的是:{}, 有{}篇'.format(categories[max_index], maxi))
```
