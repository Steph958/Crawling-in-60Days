## 作業目標

## 1. 取出今天所有的新聞

```python
# 打開瀏覽器

from selenium import webdriver
from bs4 import BeautifulSoup

browser = webdriver.Chrome(executable_path='chromedriver')

browser.get("https://www.ettoday.net/news/news-list.htm")

# 每個兩秒鐘自動往下滑

import time
from datetime import datetime as dt

flag = 0
stop = 0

for i in range(10):
    time.sleep(2)
    browser.execute_script("window.scrollBy(0, 10000);")
   
html_source = browser.page_source
soup = BeautifulSoup(html_source, "html5lib")

for d in soup.find(class_="part_list_2").find_all('h3'):
    if flag == 0:
        today = dt.strptime(d.find(class_='date').text.split(' ')[0], '%Y/%m/%d')
        flag = 1

    target = dt.strptime(d.find(class_='date').text.split(' ')[0], '%Y/%m/%d')
    if (today-target).days != 0:
        stop = 1
        break
    print(d.find(class_="date").text, d.find_all('a')[-1].text)
        
browser.quit()
```

## 2. 取出現在時間兩小時內的新聞

```python
import datetime

browser = webdriver.Chrome(executable_path='chromedriver')

browser.get("https://www.ettoday.net/news/news-list.htm")

stop = 0

now = dt.now()
delta = datetime.timedelta(hours=2)

deadlock = now - delta

for i in range(3):
    time.sleep(2)
    browser.execute_script("window.scrollBy(0, 10000);")

html_source = browser.page_source
soup = BeautifulSoup(html_source, "html5lib")

for d in soup.find(class_="part_list_2").find_all('h3'):
    target = dt.strptime(d.find(class_='date').text, '%Y/%m/%d %H:%M')
    if target < deadlock:
        break
    print(d.find(class_="date").text, d.find_all('a')[-1].text)

browser.quit()
```

## 3. 根據範例，取出三天前下午三點到五點的新聞

```python
today = dt.now()
baseday = today - datetime.timedelta(days=3)             # 設定三天前

basetime = datetime.datetime(baseday.year, baseday.month, baseday.day, 15, 0, 0)    # 設定下午三點
delta = datetime.timedelta(hours=2)                                                 # 到五點的時間差

# print(basetime)
# print(basetime+delta)

browser = webdriver.Chrome(executable_path='chromedriver')
url = "https://www.ettoday.net/news/news-list.htm"
new_url = url.replace('news-list', 'news-list-'+str(baseday.year)+'-'+str(baseday.month)+'-'+str(baseday.day)+'-0')
#print(new_url)

browser.get(new_url)

for i in range(40):
    time.sleep(2)
    browser.execute_script("window.scrollBy(0, 10000);")
    
html_source = browser.page_source
soup = BeautifulSoup(html_source, "html5lib")

for d in soup.find(class_="part_list_2").find_all('h3'):
    target = dt.strptime(d.find(class_='date').text, '%Y/%m/%d %H:%M')
#    print(d.find(class_="date").text, d.find_all('a')[-1].text)
    if target < basetime:
        break
    elif target < (basetime+delta):
        print(d.find(class_="date").text, d.find_all('a')[-1].text)
    else:
        pass
    
browser.quit()
```
