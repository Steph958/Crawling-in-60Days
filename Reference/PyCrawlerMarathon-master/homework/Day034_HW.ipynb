## 作業目標

* 目前程式中的 proxy_ips 是手動輸入的，請根據 https://free-proxy-list.net/ 寫一個可自動化抓取可用 Proxy 的 proxy_ips。




```python
import requests
from bs4 import BeautifulSoup
import random
import time
import json

requests.adapters.DEFAULT_RETRIES = 10
```

爬取 proxy list

```python
proxy_ips = []
response = requests.get('https://free-proxy-list.net/')
soup = BeautifulSoup(response.text, 'html.parser')

proxy_info = soup.find('tbody').find_all('tr')

for proxy in proxy_info:
    tmp = proxy.find_all('td')
    ip = tmp[0].text+':'+tmp[1].text    
    proxy_ips.append(ip)

print(len(proxy_ips))
print(proxy_ips)
```

驗證有效的 proxy

```python
proxy_list = []

for ip in proxy_ips:
    print('Use', ip)
    try:
        resp = requests.get('https://httpbin.org/ip', proxies={'http': ip, 'https': ip},  timeout=30, 
                            headers = {'Connection': 'close'})
        origin = json.loads(resp.text)['origin']
        print(origin)
        address = ip.split(':')[0]
        print(address)
        if address == origin:
            proxy_list.append(ip)
    except Exception as e:
        print(e)
        print()
        pass
    time.sleep(5)
```

```python
print(proxy_list)
```

再次驗證 proxy

```python
for i in range(20):
    ip = random.choice(proxy_list)
    print('Use', ip)
    try:
        resp = requests.get('https://httpbin.org/ip',
                        proxies={'http': ip, 'https': ip}, timeout=30, headers = {'Connection': 'close'})
        origin = json.loads(resp.text)['origin']
        print('Your IP address:', origin)
    except Exception as e:
        print(e)
        print()
    time.sleep(10)
```
