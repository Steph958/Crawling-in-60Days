## 作業目標

* 找一個之前實作過的爬蟲改用多線程改寫，比較前後時間的差異。





```python
import requests
from bs4 import BeautifulSoup
import random
import time
import json
import _thread

requests.adapters.DEFAULT_RETRIES = 10
```

```python
proxy_ips = []
response = requests.get('https://free-proxy-list.net/')
soup = BeautifulSoup(response.text, 'html.parser')

proxy_info = soup.find('tbody').find_all('tr')

for proxy in proxy_info:
    tmp = proxy.find_all('td')
    ip = tmp[0].text+':'+tmp[1].text    
    proxy_ips.append(ip)

print(len(proxy_ips))
print(proxy_ips)
```

```python
def verify_proxy(ip, l):
    print('Use: ', ip)
    try:
        resp = requests.get('https://httpbin.org/ip', proxies={'http': ip, 'https': ip}, headers = {'Connection': 'close'})
        origin = json.loads(resp.text)['origin']
        address = ip.split(':')[0]
        if address == origin:
            l.append(ip)
    except:
        pass
```

```python
proxy_list1 = []

start_time1 = time.time()

for ip in proxy_ips:
    verify_proxy(ip, proxy_list1)

finish_time1 = time.time()
time_origin = finish_time1 - start_time1
print('Spend: ', time_origin, ' seconds')
print(proxy_list1)
```

```python
proxy_list2 = []

start_time2 = time.time()

for ip in proxy_ips:
    _thread.start_new_thread(verify_proxy, (ip, proxy_list2,))

```

```python
time.sleep(10)
```

```python
finish_time2 = time.time()
time_multi = finish_time2 - start_time2 -10
print('Spend: ', time_multi, ' seconds')
print(proxy_list2)
```
